## Supervised Learning: Classification vs Regression.

The two main tasks handled by (supervised) ML is regression and classification.

In regression we aim at modeling the relationship between the system's response (dependent variable) and one or more explanatory variables (independent variables), acquiring a continuum of values.

Examples of regression would be predicting the temperature for each day of the year, or expenses of the household as a function of the number of children and adults.

In classification the aim is to identify what class does a data-point belong to. For example, the species or the iris plant based on the size of its petals, or whether an email is spam or not based on its content.

## Performance measures (model evaluation and training)

Performance measures evaluate how much the model reflects the dataset, and differ according to the task (Regression/classification), and to the application. Here are the most popular:


### 1. Regression

Let $\hat y(x_i)$ be the prediction for input $x_i$, $y_i$ the true value, and $\bar y=\frac{1}{n}\sum_i y_i$ the mean.

- **Mean Square Error (MSE):**
  $$\mathrm{MSE} = \frac{1}{n} \sum_i (y_i - \hat y(x_i))^2$$
- **Root Mean Square Error (RMSE):**
  $$\mathrm{RMSE} = \sqrt{\frac{1}{n} \sum_i (y_i - \hat y(x_i))^2}$$
- **Mean Absolute Error (MAE):**
  $$\mathrm{MAE} = \frac{1}{n} \sum_i |y_i - \hat y(x_i)|$$
- **Explained Variance ($R^2$):**
  $$R^2 = 1 - \frac{\sum_i (y_i - \hat y(x_i))^2}{\sum_i (y_i - \bar y)^2}$$

---

### 2. Classification

The confusion matrix and related metrics are the workhorses for classification. We distinguish binary and multiclass cases:

#### **Binary Case**

|            | Predicted 0 (Negative) | Predicted 1 (Positive) |
|------------|------------------------|------------------------|
| **True 0** | TN (True Negative)     | FP (False Positive)    |
| **True 1** | FN (False Negative)    | TP (True Positive)     |

(class 0 is the negative one, class 1 the positive)

- **Accuracy:**
  $$\mathrm{Accuracy} = \frac{TP + TN}{TP + FP + FN + TN}$$
- **Precision:**
  $$\mathrm{Precision} = \frac{TP}{TP + FP}$$
  ( For every time that I detected class 1, how many times was I correct? )
- **Recall:**
  $$\mathrm{Recall} = \frac{TP}{TP + FN}$$
  ( Among all possible samples with class 1, how many was I able to detect? )

---

#### **Multiclass Case**

The confusion matrix is a $K \times K$ table ($K$ = number of classes):

|        | Pred 0 | Pred 1 | Pred 2 | Pred 3 | Pred 4 |
|--------|--------|--------|--------|--------|--------|
| Observed class 0 | $a_{00}$ | $a_{01}$ | $a_{02}$ | $a_{03}$ | $a_{04}$ |
| Observed class 1 | $a_{10}$ | $a_{11}$ | $a_{12}$ | $a_{13}$ | $a_{14}$ |
| Observed class 2 | $a_{20}$ | $a_{21}$ | $a_{22}$ | $a_{23}$ | $a_{24}$ |
| Observed class 3 | $a_{30}$ | $a_{31}$ | $a_{32}$ | $a_{33}$ | $a_{34}$ |
| Observed class 4 | $a_{40}$ | $a_{41}$ | $a_{42}$ | $a_{43}$ | $a_{44}$ |

- **Accuracy:**
  $$\mathrm{Accuracy} = \frac{\sum_{i=0}^K a_{ii}}{\sum_{i,j=0}^K a_{ij}}$$
- **Precision, Recall:**
  We can define precision and recall *per class*. Try to answer the questions: ( *For every time that I detected class xxx, how many times was I correct?* , and *Among all possible samples with class xxx, how many was I able to detect?* )


Precision = Prediction (a33)/ sum of Predictions (a03 to a43)

We get one precision per class. So some classes are more precise and some or less.


A similarity measure between the model and the true dataset is needed both for:

- **training/fitting a model (loss function)** 
- **performance evaluation (metrics)**.

For training, usually **differentiable loss functions** are employed, which can be minimized via gradient descent. For regression tasks, common choices are the Mean Square Error (MSE) or the Mean Absolute Error (MAE). For classification tasks, cross-entropy loss is often used.

## Actual aim: Generalization

In supervised machine learning we distinguish two faces: `training` and `testing`. To measure model performance in an unbiased way, we need to use different data to test the model than the data that the model was trained on. 

Therefore we often use the 'train-test' split: e.g. 20% of all available dataset is reserved for model performance test, and the remaining 80% is used for actual model training (these percentages may vary).

CAVEAT: When the test set contains information about the training set that can spoil the resulting metrics we talk about `data leakage` or `information leakage`.

### Optimizing generalization error:

To understand the origins of poor generalization we need to identify two source of errors: bias and variance. In order to understand this concept we perform a thought experiment. Suppose we use N examples to fit our model, and get a prediction $f_1(x)$. Now we choose other N examples and get another prediction $f_2(x)$ for the same sample. We can follow this procedure many times collecting $f_i(x),i=1...N_{trials}$.

The **variance** of the model is the spread of $f_i(x), i=1...N_{trials}$, saying how much the predictions are stable.

The **bias** of the model compares the mean value of $f_i(x), i=1...N_{trials}$ w.r.t to the real value $f(x$)$.

The total error has a bias and a variance component.

```python
import os
import tensorflow as tf
import tarfile
import os
from packaging import version
import shutil
import sys 


def download_and_extract_data(
    url="https://github.com/neworldemancer/DSF5/raw/master/colab_material.tgz",
    target_dir="data",
    fname="colab_material.tgz",
    update_folder=False
):
    """Download and extract a tar.gz dataset into target_dir."""
    
    if update_folder and os.path.exists(target_dir):
        shutil.rmtree(target_dir)

    if not os.path.exists(target_dir):
        cache_dir = os.path.abspath(".")

        if version.parse(tf.__version__) >= version.parse("2.13.0"):
            # new behavior: fname must be only a filename
            path = tf.keras.utils.get_file(
                fname=fname,
                origin=url,
                cache_dir=cache_dir
            )
        else:
            # old behavior: can pass full path
            path = tf.keras.utils.get_file(
                fname=os.path.join(cache_dir, fname),
                origin=url
            )
        # extract tar into target_dir
        with tarfile.open(path, "r:gz") as tar:
            tar.extractall(target_dir)
    else:
        print('Data already present. Use update_folder = True to overwrite/update if desired.')
    return os.path.abspath(target_dir)

data_path = download_and_extract_data(update_folder=False)
sys.path.append(data_path)
print("Data available at:", data_path)

import utils.routines as routines
```

## 1. Synthetic linear

```python
def get_linear(n_d=1, n_points=10, w=None, b=None, sigma=5):
  x = np.random.uniform(0, 10, size=(n_points, n_d))

  w = w or np.random.uniform(0.1, 10, n_d)
  b = b or np.random.uniform(-10, 10)
  y = np.dot(x, w) + b + np.random.normal(0, sigma, size=n_points)

  print('true slopes: w =', w, ';  b =', b)

  return x, y
```

```python
x, y = get_linear(n_d=1, sigma=1)
plt.plot(x[:, 0], y, '*')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

The module contains A LOT of different machine learning methods, and here we will cover only few of them. What is great about `scikit-learn` is that it has a uniform and consistent interface.

All the different ML approaches are implemented as classes with a set of same main methods:

1. `fitter = ...`: Create fitter object.
2. `fitter.fit(x, y[, sample_weight])`: Fit model to predict from list of smaples `x` a list of target values `y`.
3. `y_pred = fitter.predict(X)`: Predict using the trained model.
4. `s = fitter.score(x, y[, sample_weight])`: Obtain a relevant performance measure of the trained model.

This allows one to easily replace one approach with another and find the best one for the problem at hand, by simply using a regression/classification object of another class, while the rest of the code can remain the same.


It is useful to know that generally in scikit-learn the input data is represented as a matrix $X$ of dimensions `n_samples x n_features` (also called the `design matrix`), whereas the supervised labels/values are stored in a matrix $Y$ of dimensions `n_samples x n_target` .

# 1.Linear models

In many cases the scalar value of interest - dependent variable - is (or can be approximated as) linear combination of the independent variables.

In linear regression the estimator is searched in the form: $$\hat{y}(\bar{x} | w_0,\bar{w}) = w_0 + w_1 x_1 + ... + w_p x_p$$

The parameters $\bar{w} = (w_1,..., w_p)$ and $w_0$ are designated as `coef_` and `intercept_` in `sklearn`.

Reference: https://scikit-learn.org/stable/modules/linear_model.html

## 1. Linear regression

[LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) fits a linear model with coefficients $\bar{w} = (w_1,..., w_p)$ and $w_0$ to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.

Mathematically it solves a problem of the form:  $$(w^{opt}_0,\bar{w}^{opt}) = \arg min_{w_0,\bar{w}} \sum_i \left(\hat{y}(\bar{x_i} | w_0,\bar{w})-y_i\right)^2$$

The function:

$$L(w,\bar{w}) \equiv \sum_i \left(\hat{y}(\bar{x_i} | w_0,\bar{w})-y_i\right)^2$$

is called the `loss function` for linear regression and we say that during training/fitting the loss function is `minimized`.

In terms of the design matrix $X_{i,p}$, where $i$ ranges over the number of samples and $p$ across the number of features:

$$L(w,\bar{w})=\sum_i \left((w_0 + \sum_p X_{i,p}w_p)-y_i\right)^2$$

This makes it explicit that the function to minimize is quadratic in $w$ and an analytical solution is therefore available (no numerical optimization).

```python
x, y = get_linear(n_d=1, sigma=2, n_points=1000)  # p==1, 1D input
plt.scatter(x, y);
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```

## 2. Logistic regression

## Logistic Regression: Binary Case

Logistic regression is a **linear model for classification**, not regression. Typical binary outputs include:
- {sold, not sold} (e.g., predicting if a product will be sold)
- {failed, passed} (e.g., predicting if a student will fail a class)

We map the two classes to targets $y_i \in \{0,1\}$, called the positive and negative class, respectively.

**Model:**
The probability $p$ of a point belonging to the positive class is modeled as:
$$p = \sigma(w_0 + w_1 x_1 + ... + w_p x_p)$$
where the sigmoid function is:
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

The sigmoid maps the real-valued output of a linear function to $[0,1]$.

**Loss function:**
Parameters $w$ are trained to minimize:
$$L(w_0, \bar{w}) = -\sum_i y_i \log(p_i)$$

## Logistic Regression: More Classes

A **multiclass** model assigns each sample to one of several mutually exclusive classes (e.g., predicting treatment response: {negative, not significant, positive}).

To extend the binary model, we use **multinomial logistic regression**. This computes logits for each class and aggregates them using the softmax operator:

$$
\text{logits}_c = w_{c,0} + w_{c,1} x_1 + ... + w_{c,p} x_p
$$
$$
p_c = \text{softmax}(\text{logits})_c = \frac{e^{\text{logits}_c}}{\sum_{c'} e^{\text{logits}_{c'}}}
$$

Probabilities are normalized before training.

**Loss function:**
Parameters are trained to minimize:
$$
L(\{w_{0,c}\}, \{\bar{w}_c\}) = -\sum_i \left(\sum_c y_i^{OH}(c) \log p_{i,c}\right)
$$
where $y_i^{OH}(c) = 1$ if sample $i$ belongs to class $c$, $0$ otherwise (one-hot encoding).

This converts a multiclass label (e.g., $y_i \in \{cat, dog, other\}$) into binary labels: $y_i^{OH}(dog), y_i^{OH}(cat), y_i^{OH}(other) \in \{0,1\}$.

---
**Notes:**
- The choices here are based on probability theory and maximum likelihood.
- If classes are not exclusive (multilabel), fit individual binary models (one-vs-rest, `ovr`).

You can use the following function that takes the frequency content of the feature (power spectrum) as an example.

The process of trying out several features extracted from the data as input to a model is called *(manual) feature engineering*

```python
#Helper function for point 2

def extract_features(images):
    """
    Extracts power spectrum features from a list or array of images.

    For each image, computes the 2D Fourier transform, calculates the power spectrum,
    normalizes it, and flattens the result into a 1D feature vector.

    Args:
        images (array-like): List or array of 2D image arrays.

    Returns:
        np.ndarray: Array of shape (num_images, num_features) containing the power spectrum features for each image.
    """
    def compute_power_spectrum(image):
        fourier_transform = np.fft.fft2(image)
        power_spectrum = np.abs(fourier_transform) ** 2
        power_spectrum /= np.sum(power_spectrum)
        return power_spectrum.flatten()
    power_spectrum_features = []
    for img in images:
        spectrum = compute_power_spectrum(img)
        power_spectrum_features.append(spectrum)
    return np.array(power_spectrum_features)
```

We can summarize the confusion matrix plotting precision and recall for each category. Given the confusion matrix you can use the following funtion:

```python
# Helper function for point 6

def get_class_specific_prec_recall(cm):
    """
    Given a confusion matrix cm, computes precision and recall for each class.

    Precision for class i: cm[i, i] / sum of column i (true positives / predicted positives)
    Recall for class i: cm[i, i] / sum of row i (true positives / actual positives)

    Args:
        cm (np.ndarray): Confusion matrix of shape (n_classes, n_classes)

    Returns:
        precisions (list): Precision for each class
        recalls (list): Recall for each class
    """
    n_cat = cm.shape[0]
    precisions = []
    recalls = []
    # precision
    for i in range(n_cat):
        precisions.append(cm[i, i] / np.sum(cm[:, i]))

    # recall
    for i in range(n_cat):
        recalls.append(cm[i, i] / np.sum(cm[i, :]))
    return precisions, recalls
```

# Visualizing Multiclass Logistic Regression Training Process

There are several ways to visualize how multiclass logistic regression learns on high-dimensional data like Fashion-MNIST:

1. **Loss function progression**: See how the cross-entropy loss decreases over iterations
2. **Class probability evolution**: Watch how predicted probabilities for each class change during training
3. **Weight evolution**: Observe how the model weights for different classes change
4. **Feature importance evolution**: See how different features become important for classification
5. **Confusion matrix evolution**: Track how classification accuracy improves over time

Let's implement these visualizations for the Fashion-MNIST dataset!

```python
# Custom Multiclass Logistic Regression with Training Visualization
class VisualizableMulticlassLogisticRegression:
    def __init__(self, learning_rate=0.01, max_iter=1000, n_classes=10):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.n_classes = n_classes
        self.losses = []
        self.weights_history = []
        self.accuracy_history = []
        self.class_probabilities_history = []
        
    def softmax(self, z):
        # Numerically stable softmax
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)
    
    def one_hot_encode(self, y):
        one_hot = np.zeros((len(y), self.n_classes))
        one_hot[np.arange(len(y)), y] = 1
        return one_hot
    
    def fit(self, X, y, store_every=10):
        # Add intercept term
        X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
        
        # Initialize weights randomly (one set per class)
        self.weights = np.random.normal(0, 0.01, (X_with_intercept.shape[1], self.n_classes))
        
        # One-hot encode labels
        y_onehot = self.one_hot_encode(y)
        
        for i in range(self.max_iter):
            # Forward pass
            z = X_with_intercept.dot(self.weights)  # (n_samples, n_classes)
            predictions = self.softmax(z)  # (n_samples, n_classes)
            
            # Compute loss (cross-entropy)
            epsilon = 1e-15
            predictions = np.clip(predictions, epsilon, 1 - epsilon)
            loss = -np.mean(np.sum(y_onehot * np.log(predictions), axis=1))
            
            # Compute accuracy
            predicted_classes = np.argmax(predictions, axis=1)
            accuracy = np.mean(predicted_classes == y)
            
            # Store metrics
            if i % store_every == 0:
                self.losses.append(loss)
                self.weights_history.append(self.weights.copy())
                self.accuracy_history.append(accuracy)
                # Store probabilities for a few sample points
                sample_indices = np.random.choice(len(y), min(20, len(y)), replace=False)
                self.class_probabilities_history.append(predictions[sample_indices].copy())
            
            # Compute gradients
            gradient = X_with_intercept.T.dot(predictions - y_onehot) / X.shape[0]
            
            # Update weights
            self.weights -= self.learning_rate * gradient
            
            # Early stopping if loss is very small
            if loss < 1e-6:
                break
                
        # Final storage
        if len(self.losses) == 0 or self.losses[-1] != loss:
            self.losses.append(loss)
            self.weights_history.append(self.weights.copy())
            self.accuracy_history.append(accuracy)
            sample_indices = np.random.choice(len(y), min(20, len(y)), replace=False)
            final_predictions = self.softmax(X_with_intercept.dot(self.weights))
            self.class_probabilities_history.append(final_predictions[sample_indices].copy())
    
    def predict_proba(self, X):
        X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
        return self.softmax(X_with_intercept.dot(self.weights))
    
    def predict(self, X):
        return np.argmax(self.predict_proba(X), axis=1)
```

```python
# Extract weight evolution for multiclass model
weights_array = np.array(model.weights_history)  # Shape: (n_iterations, n_features, n_classes)
iterations = np.arange(len(weights_array)) * 5  # Since we stored every 5th iteration

print(f"Weights shape: {weights_array.shape} (iterations, features, classes)")

plt.figure(figsize=(16, 12))

# Plot 1: Weight magnitudes for each class over time
plt.subplot(3, 2, 1)
for class_idx in range(10):
    # Calculate L2 norm of weights for each class (excluding intercept)
    weight_norms = np.sqrt(np.sum(weights_array[:, 1:, class_idx]**2, axis=1))
    plt.plot(iterations, weight_norms, label=f'{fmnist_class_names[class_idx]}', linewidth=2)
plt.xlabel('Training Iteration')
plt.ylabel('Weight L2 Norm')
plt.title('Weight Magnitude Evolution per Class')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.grid(True, alpha=0.3)

# Plot 2: Intercept (bias) evolution for each class
plt.subplot(3, 2, 2)
for class_idx in range(10):
    intercepts = weights_array[:, 0, class_idx]  # First weight is intercept
    plt.plot(iterations, intercepts, label=f'{fmnist_class_names[class_idx]}', linewidth=2)
plt.xlabel('Training Iteration')
plt.ylabel('Intercept Value')
plt.title('Intercept Evolution per Class')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.grid(True, alpha=0.3)

# Plot 3: Weight change rate (gradient magnitude) over time
plt.subplot(3, 2, 3)
if len(weights_array) > 1:
    weight_changes = np.diff(weights_array, axis=0)
    change_magnitudes = np.sqrt(np.sum(weight_changes**2, axis=(1, 2)))  # Total change magnitude
    plt.plot(iterations[1:], change_magnitudes, 'purple', linewidth=2)
    plt.xlabel('Training Iteration')
    plt.ylabel('Total Weight Change Magnitude')
    plt.title('Rate of Weight Change (Learning Speed)')
    plt.grid(True, alpha=0.3)

# Plot 4: Feature importance evolution (sum of absolute weights across classes)
plt.subplot(3, 2, 4)
# Show evolution of top 10 most important features
feature_importance_evolution = np.sum(np.abs(weights_array[:, 1:, :]), axis=2)  # Sum across classes, exclude intercept
top_features = np.argsort(feature_importance_evolution[-1])[-10:]  # Top 10 at final iteration

for feature_idx in top_features:
    plt.plot(iterations, feature_importance_evolution[:, feature_idx], 
             label=f'Feature {feature_idx}', alpha=0.8)
plt.xlabel('Training Iteration')
plt.ylabel('Feature Importance (Sum |weights|)')
plt.title('Evolution of Top Feature Importances')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.grid(True, alpha=0.3)

# Plot 5: Weight distribution comparison (initial vs final)
plt.subplot(3, 2, 5)
initial_weights = weights_array[0, 1:, :].flatten()  # Exclude intercept
final_weights = weights_array[-1, 1:, :].flatten()   # Exclude intercept

plt.hist(initial_weights, bins=50, alpha=0.5, label='Initial weights', density=True)
plt.hist(final_weights, bins=50, alpha=0.5, label='Final weights', density=True)
plt.xlabel('Weight Value')
plt.ylabel('Density')
plt.title('Weight Distribution: Initial vs Final')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 6: Class separability (weight differences)
plt.subplot(3, 2, 6)
# Calculate pairwise differences in weight norms between classes
final_weight_norms = np.sqrt(np.sum(weights_array[-1, 1:, :]**2, axis=0))
plt.bar(range(10), final_weight_norms, color=plt.cm.tab10(range(10)), alpha=0.8)
plt.xlabel('Class')
plt.ylabel('Final Weight Norm')
plt.title('Final Weight Magnitude by Class')
plt.xticks(range(10), [name[:4] for name in fmnist_class_names], rotation=45)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print some statistics
print("\nWeight Evolution Statistics:")
print(f"Initial total weight magnitude: {np.sqrt(np.sum(weights_array[0]**2)):.4f}")
print(f"Final total weight magnitude: {np.sqrt(np.sum(weights_array[-1]**2)):.4f}")
print(f"Average weight change per iteration: {np.mean(change_magnitudes):.6f}")

# Show which features became most important
final_feature_importance = np.sum(np.abs(weights_array[-1, 1:, :]), axis=1)
top_10_features = np.argsort(final_feature_importance)[-10:]
print(f"\nTop 10 most important features (by total absolute weight): {top_10_features}")
```

## 4. Confusion Matrix Evolution

Let's see how the classification performance improves during training by tracking the confusion matrix:

```python
# Create confusion matrices at different training stages
def compute_confusion_matrix_at_iteration(weights, X, y_true):
    """Compute confusion matrix using weights from a specific iteration"""
    X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
    z = X_with_intercept.dot(weights)
    predictions = np.exp(z - np.max(z, axis=1, keepdims=True))
    predictions = predictions / np.sum(predictions, axis=1, keepdims=True)  # Softmax
    y_pred = np.argmax(predictions, axis=1)
    
    cm = np.zeros((10, 10), dtype=int)
    for true_class in range(10):
        for pred_class in range(10):
            cm[true_class, pred_class] = np.sum((y_true == true_class) & (y_pred == pred_class))
    
    return cm, y_pred

# Select key iterations to show confusion matrices
key_iterations = [0, len(model.weights_history)//4, len(model.weights_history)//2, -1]
iteration_labels = ['Initial', 'Early Training', 'Mid Training', 'Final']

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()

cms = []
accuracies = []

for i, iter_idx in enumerate(key_iterations):
    weights = model.weights_history[iter_idx]
    cm, y_pred = compute_confusion_matrix_at_iteration(weights, X_vis, y_vis)
    cms.append(cm)
    
    accuracy = np.trace(cm) / np.sum(cm)
    accuracies.append(accuracy)
    
    # Plot confusion matrix
    ax = axes[i]
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.set_title(f'{iteration_labels[i]} (Accuracy: {accuracy:.3f})')
    
    # Add text annotations
    thresh = cm.max() / 2.
    for row in range(10):
        for col in range(10):
            ax.text(col, row, cm[row, col], ha="center", va="center",
                   color="white" if cm[row, col] > thresh else "black", fontsize=8)
    
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('True Label')
    ax.set_xticks(range(10))
    ax.set_yticks(range(10))
    ax.set_xticklabels([name[:4] for name in fmnist_class_names], rotation=45)
    ax.set_yticklabels([name[:4] for name in fmnist_class_names])

plt.tight_layout()
plt.show()

# Plot evolution of per-class precision and recall
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

precisions_evolution = []
recalls_evolution = []

for cm in cms:
    precisions = []
    recalls = []
    for class_idx in range(10):
        # Precision: TP / (TP + FP)
        tp = cm[class_idx, class_idx]
        fp = np.sum(cm[:, class_idx]) - tp
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        precisions.append(precision)
        
        # Recall: TP / (TP + FN)
        fn = np.sum(cm[class_idx, :]) - tp
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        recalls.append(recall)
    
    precisions_evolution.append(precisions)
    recalls_evolution.append(recalls)

precisions_evolution = np.array(precisions_evolution)
recalls_evolution = np.array(recalls_evolution)

# Plot precision evolution
for class_idx in range(10):
    ax1.plot(range(len(key_iterations)), precisions_evolution[:, class_idx], 
             'o-', label=fmnist_class_names[class_idx], linewidth=2, markersize=6)

ax1.set_xlabel('Training Stage')
ax1.set_ylabel('Precision')
ax1.set_title('Precision Evolution by Class')
ax1.set_xticks(range(len(key_iterations)))
ax1.set_xticklabels(iteration_labels)
ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax1.grid(True, alpha=0.3)
ax1.set_ylim(0, 1)

# Plot recall evolution
for class_idx in range(10):
    ax2.plot(range(len(key_iterations)), recalls_evolution[:, class_idx], 
             'o-', label=fmnist_class_names[class_idx], linewidth=2, markersize=6)

ax2.set_xlabel('Training Stage')
ax2.set_ylabel('Recall')
ax2.set_title('Recall Evolution by Class')
ax2.set_xticks(range(len(key_iterations)))
ax2.set_xticklabels(iteration_labels)
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax2.grid(True, alpha=0.3)
ax2.set_ylim(0, 1)

plt.tight_layout()
plt.show()

# Print final performance summary
print("\n=== Final Training Results ===")
print(f"Final Accuracy: {accuracies[-1]:.3f}")
print(f"Improvement from initial: {accuracies[-1] - accuracies[0]:.3f}")

print("\nFinal Per-Class Performance:")
final_precisions = precisions_evolution[-1]
final_recalls = recalls_evolution[-1]

for i, class_name in enumerate(fmnist_class_names):
    f1_score = 2 * final_precisions[i] * final_recalls[i] / (final_precisions[i] + final_recalls[i]) if (final_precisions[i] + final_recalls[i]) > 0 else 0
    print(f"{class_name:12}: Precision={final_precisions[i]:.3f}, Recall={final_recalls[i]:.3f}, F1={f1_score:.3f}")

# Find most/least confused classes
final_cm = cms[-1]
off_diagonal = final_cm.copy()
np.fill_diagonal(off_diagonal, 0)
max_confusion_idx = np.unravel_index(np.argmax(off_diagonal), off_diagonal.shape)
max_confusion_value = off_diagonal[max_confusion_idx]

print(f"\nMost confused pair: {fmnist_class_names[max_confusion_idx[0]]}  {fmnist_class_names[max_confusion_idx[1]]} ({max_confusion_value} misclassifications)")
```

```python
# Create animated version of multiclass training evolution
from matplotlib.animation import FuncAnimation

def create_multiclass_training_animation():
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    def animate(frame):
        # Clear all axes
        for ax in [ax1, ax2, ax3, ax4]:
            ax.clear()
        
        # Get current iteration data
        current_frame = min(frame, len(model.losses) - 1)
        current_loss = model.losses[current_frame]
        current_accuracy = model.accuracy_history[current_frame]
        
        # Plot 1: Loss evolution up to current frame
        ax1.plot(range(current_frame + 1), model.losses[:current_frame + 1], 'b-', linewidth=2)
        ax1.scatter(current_frame, current_loss, c='red', s=100, zorder=5)
        ax1.set_xlabel('Training Iteration')
        ax1.set_ylabel('Cross-Entropy Loss')
        ax1.set_title(f'Loss Evolution - Current: {current_loss:.4f}')
        ax1.grid(True, alpha=0.3)
        ax1.set_xlim(0, len(model.losses))
        ax1.set_ylim(0, max(model.losses) * 1.1)
        
        # Plot 2: Accuracy evolution up to current frame  
        ax2.plot(range(current_frame + 1), model.accuracy_history[:current_frame + 1], 'g-', linewidth=2)
        ax2.scatter(current_frame, current_accuracy, c='red', s=100, zorder=5)
        ax2.set_xlabel('Training Iteration')
        ax2.set_ylabel('Accuracy')
        ax2.set_title(f'Accuracy Evolution - Current: {current_accuracy:.3f}')
        ax2.grid(True, alpha=0.3)
        ax2.set_xlim(0, len(model.accuracy_history))
        ax2.set_ylim(0, 1)
        
        # Plot 3: Weight magnitude evolution for each class
        if current_frame > 0:
            weights_up_to_now = np.array(model.weights_history[:current_frame + 1])
            for class_idx in range(min(5, 10)):  # Show first 5 classes to avoid clutter
                weight_norms = np.sqrt(np.sum(weights_up_to_now[:, 1:, class_idx]**2, axis=1))
                ax3.plot(range(current_frame + 1), weight_norms, 
                        label=f'{fmnist_class_names[class_idx][:6]}', linewidth=2)
        ax3.set_xlabel('Training Iteration')
        ax3.set_ylabel('Weight L2 Norm')
        ax3.set_title('Weight Magnitude Evolution (First 5 Classes)')
        ax3.legend(fontsize=8)
        ax3.grid(True, alpha=0.3)
        
        # Plot 4: Current class probabilities for first few samples
        if current_frame < len(model.class_probabilities_history):
            current_probs = model.class_probabilities_history[current_frame]
            n_samples_to_show = min(3, current_probs.shape[0])
            
            x_pos = np.arange(10)
            width = 0.25
            
            for sample_idx in range(n_samples_to_show):
                ax4.bar(x_pos + sample_idx * width, current_probs[sample_idx], 
                       width, label=f'Sample {sample_idx + 1}', alpha=0.8)
            
            ax4.set_xlabel('Class')
            ax4.set_ylabel('Predicted Probability')
            ax4.set_title(f'Current Class Probabilities (Iteration {current_frame * 5})')
            ax4.set_xticks(x_pos + width)
            ax4.set_xticklabels([name[:4] for name in fmnist_class_names], rotation=45)
            ax4.legend(fontsize=8)
            ax4.grid(True, alpha=0.3)
            ax4.set_ylim(0, 1)
        
        # Add iteration info
        fig.suptitle(f'Multiclass Logistic Regression Training - Iteration {current_frame * 5}', 
                    fontsize=16, y=0.98)
    
    # Create animation
    anim = FuncAnimation(fig, animate, frames=len(model.losses), interval=300, repeat=True)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    return anim

# Create comprehensive training summary visualization
def create_training_summary():
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    
    # Plot 1: Loss vs Accuracy
    ax = axes[0]
    ax2 = ax.twinx()
    line1 = ax.plot(model.losses, 'b-', linewidth=2, label='Loss')
    line2 = ax2.plot(model.accuracy_history, 'g-', linewidth=2, label='Accuracy')
    ax.set_xlabel('Training Iteration')
    ax.set_ylabel('Cross-Entropy Loss', color='b')
    ax2.set_ylabel('Accuracy', color='g')
    ax.set_title('Training Progress Overview')
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax.legend(lines, labels)
    ax.grid(True, alpha=0.3)
    
    # Plot 2: Final confusion matrix
    if 'cms' in locals() and len(cms) > 0:
        final_cm = cms[-1]
        im = axes[1].imshow(final_cm, interpolation='nearest', cmap='Blues')
        axes[1].set_title('Final Confusion Matrix')
        axes[1].set_xlabel('Predicted Label')
        axes[1].set_ylabel('True Label')
        tick_marks = np.arange(10)
        axes[1].set_xticks(tick_marks)
        axes[1].set_yticks(tick_marks)
        axes[1].set_xticklabels([name[:4] for name in fmnist_class_names], rotation=45)
        axes[1].set_yticklabels([name[:4] for name in fmnist_class_names])
        plt.colorbar(im, ax=axes[1])
    
    # Plot 3: Weight evolution heatmap
    weights_array = np.array(model.weights_history)
    # Show weight changes for first 20 features across classes
    feature_subset = weights_array[:, 1:21, :]  # First 20 features, exclude intercept
    weight_changes = np.abs(feature_subset[-1] - feature_subset[0])  # Initial vs final
    im2 = axes[2].imshow(weight_changes.T, aspect='auto', cmap='YlOrRd')
    axes[2].set_title('Feature Weight Changes\n(Initial vs Final)')
    axes[2].set_xlabel('Feature Index')
    axes[2].set_ylabel('Class')
    axes[2].set_yticks(range(10))
    axes[2].set_yticklabels([name[:4] for name in fmnist_class_names])
    plt.colorbar(im2, ax=axes[2])
    
    # Plot 4: Learning curve (loss reduction rate)
    if len(model.losses) > 1:
        loss_diff = -np.diff(model.losses)  # Positive values = loss decrease
        axes[3].plot(loss_diff, 'r-', linewidth=2)
        axes[3].set_xlabel('Training Iteration')
        axes[3].set_ylabel('Loss Reduction')
        axes[3].set_title('Learning Rate (Loss Reduction per Iteration)')
        axes[3].grid(True, alpha=0.3)
        axes[3].axhline(y=0, color='black', linestyle='--', alpha=0.5)
    
    # Plot 5: Class probability evolution summary
    if len(model.class_probabilities_history) > 0:
        prob_history = np.array(model.class_probabilities_history)
        # Average max probability (confidence) evolution
        max_probs_avg = np.mean(np.max(prob_history, axis=2), axis=1)
        axes[4].plot(max_probs_avg, 'purple', linewidth=2)
        axes[4].set_xlabel('Training Iteration')
        axes[4].set_ylabel('Average Max Probability')
        axes[4].set_title('Model Confidence Evolution')
        axes[4].grid(True, alpha=0.3)
        axes[4].set_ylim(0, 1)
    
    # Plot 6: Final model statistics
    axes[5].axis('off')
    stats_text = f"""
    Final Training Statistics:
    
     Initial Loss: {model.losses[0]:.4f}
     Final Loss: {model.losses[-1]:.4f}
     Loss Reduction: {model.losses[0] - model.losses[-1]:.4f}
    
     Initial Accuracy: {model.accuracy_history[0]:.3f}
     Final Accuracy: {model.accuracy_history[-1]:.3f}
     Accuracy Gain: {model.accuracy_history[-1] - model.accuracy_history[0]:.3f}
    
     Total Iterations: {len(model.losses)}
     Features Used: {X_vis.shape[1]}
     Classes: {len(np.unique(y_vis))}
     Training Samples: {len(y_vis)}
    """
    axes[5].text(0.1, 0.9, stats_text, transform=axes[5].transAxes, 
                verticalalignment='top', fontsize=11, 
                bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    
    plt.tight_layout()
    plt.show()

# Show the training summary
create_training_summary()

# Note: Uncomment to create and show animation (may be slow)
print("\nTo see the animated training visualization:")
print("# anim = create_multiclass_training_animation()")
print("# plt.show()")
print("# Or to save as gif:")
print("# anim.save('multiclass_training.gif', writer='pillow', fps=2)")
```
